```python
# 2. Data Cleaning
# Cleaning goal_scorers dataset
# Checking for missing values
missing_values_gs = goal_scorers.isnull().sum()
print("\nMissing Values in Goal Scorers Dataset:\n", missing_values_gs)
```

```python
# Drop rows with missing values in 'scorer' and 'minute' columns
goal_scorers_cleaned = goal_scorers.dropna(subset=['scorer', 'minute'])
print("\nRows before dropping missing values:", len(goal_scorers))
print("Rows after dropping missing values:", len(goal_scorers_cleaned))
```

```python
# Checking for duplicate rows
duplicate_rows_gs = goal_scorers_cleaned.duplicated().sum()
print("Duplicate Rows in Goal Scorers Dataset:", duplicate_rows_gs)
```

```python
# Drop duplicate rows
goal_scorers_cleaned = goal_scorers_cleaned.drop_duplicates()
print("Rows after dropping duplicates:", len(goal_scorers_cleaned))
```

```python
# Convert 'date' column to datetime type
goal_scorers_cleaned['date'] = pd.to_datetime(goal_scorers_cleaned['date'])
```

```python
# Verifying data types
print("\nGoal Scorers Dataset - Data Types after Cleaning:")
print(goal_scorers_cleaned.dtypes)
```

```python
# Cleaning match_results dataset
# Checking for missing values
missing_values_res = match_results.isnull().sum()
print("\nMissing Values in Match Results Dataset:\n", missing_values_res)
```

```python
# Drop rows with missing values
match_results_cleaned = match_results.dropna()
print("\nRows before dropping missing values:", len(match_results))
print("Rows after dropping missing values:", len(match_results_cleaned))
```

```python
# Checking for duplicate rows
duplicate_rows_res = match_results_cleaned.duplicated().sum()
print("Duplicate Rows in Match Results Dataset:", duplicate_rows_res)
```

```python
# Drop duplicate rows
match_results_cleaned = match_results_cleaned.drop_duplicates()
print("Rows after dropping duplicates:", len(match_results_cleaned))
```

```python
# Convert 'date' column to datetime type
match_results_cleaned['date'] = pd.to_datetime(match_results_cleaned['date'])
```

```python
# Verifying data types
print("\nMatch Results Dataset - Data Types after Cleaning:")
print(match_results_cleaned.dtypes)
```

```python
# Cleaning penalty_shootouts dataset
# Checking for missing values
missing_values_shoot = penalty_shootouts.isnull().sum()
print("\nMissing Values in Penalty Shootouts Dataset:\n", missing_values_shoot)
```

```python
# Drop duplicate rows
penalty_shootouts_cleaned = penalty_shootouts_cleaned.drop_duplicates()
print("Rows after dropping duplicates:", len(penalty_shootouts_cleaned))
```

```python
# Convert 'date' column to datetime type
penalty_shootouts_cleaned['date'] = pd.to_datetime(penalty_shootouts_cleaned['date'])
```

```python
# 4. Data Distribution
# Distribution of goals scored minute
sns.histplot(goal_scorers_cleaned['minute'], kde=True)
plt.title('Distribution of Goal Scoring Minutes')
plt.xlabel('Minute')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of home scores
sns.histplot(match_results_cleaned['home_score'], kde=True)
plt.title('Distribution of Home Scores')
plt.xlabel('Home Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of away scores
sns.histplot(match_results_cleaned['away_score'], kde=True)
plt.title('Distribution of Away Scores')
plt.xlabel('Away Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. Trends over Time
# Number of matches over time
match_results_cleaned['year'] = match_results_cleaned['date'].dt.year
matches_per_year = match_results_cleaned.groupby('year').size()
matches_per_year.plot()
plt.title('Number of Matches per Year')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.show()
```

```python
# Average home and away scores over time
avg_scores_per_year = match_results_cleaned.groupby('year')[['home_score', 'away_score']].mean()
avg_scores_per_year.plot()
plt.title('Average Home and Away Scores per Year')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.show()
```

```python
# Total goals scored over time
goals_per_year = goal_scorers_cleaned.groupby(goal_scorers_cleaned['date'].dt.year).size()
goals_per_year.plot()
plt.title('Total Goals per Year')
plt.xlabel('Year')
plt.ylabel('Total Goals')
plt.show()
```

```python
# Average goals per match over time
avg_goals_per_match_per_year = goals_per_year / matches_per_year
avg_goals_per_match_per_year.plot()
plt.title('Average Goals per Match per Year')
plt.xlabel('Year')
plt.ylabel('Average Goals per Match')
plt.show()
```

```python
# Outlier Detection
# Box plot for minute in goal_scorers_cleaned
sns.boxplot(x=goal_scorers_cleaned['minute'])
plt.title('Box Plot of Goal Scoring Minutes')
plt.xlabel('Minute')
plt.show()
```

```python
# Box plot for home_score in match_results_cleaned
sns.boxplot(x=match_results_cleaned['home_score'])
plt.title('Box Plot of Home Scores')
plt.xlabel('Home Score')
plt.show()
```

```python
# Box plot for away_score in match_results_cleaned
sns.boxplot(x=match_results_cleaned['away_score'])
plt.title('Box Plot of Away Scores')
plt.xlabel('Away Score')
plt.show()
```

```python
# Identifying Outliers using IQR Method
# For goal_scorers_cleaned minute
Q1_minute = goal_scorers_cleaned['minute'].quantile(0.25)
Q3_minute = goal_scorers_cleaned['minute'].quantile(0.75)
IQR_minute = Q3_minute - Q1_minute
outliers_minute = goal_scorers_cleaned[(goal_scorers_cleaned['minute'] < (Q1_minute - 1.5 * IQR_minute)) | (goal_scorers_cleaned['minute'] > (Q3_minute + 1.5 * IQR_minute))]
print("Outliers in Goal Scoring Minutes:\n", outliers_minute)
```

```python
# For match_results_cleaned home_score
Q1_home_score = match_results_cleaned['home_score'].quantile(0.25)
Q3_home_score = match_results_cleaned['home_score'].quantile(0.75)
IQR_home_score = Q3_home_score - Q1_home_score
outliers_home_score = match_results_cleaned[(match_results_cleaned['home_score'] < (Q1_home_score - 1.5 * IQR_home_score)) | (match_results_cleaned['home_score'] > (Q3_home_score + 1.5 * IQR_home_score))]
print("Outliers in Home Scores:\n", outliers_home_score)
```

```python
# For match_results_cleaned away_score
Q1_away_score = match_results_cleaned['away_score'].quantile(0.25)
Q3_away_score = match_results_cleaned['away_score'].quantile(0.75)
IQR_away_score = Q3_away_score - Q1_away_score
outliers_away_score = match_results_cleaned[(match_results_cleaned['away_score'] < (Q1_away_score - 1.5 * IQR_away_score)) | (match_results_cleaned['away_score'] > (Q3_away_score + 1.5 * IQR_away_score))]
print("Outliers in Away Scores:\n", outliers_away_score)
```

```python
# Handling Outliers by Log Transformation
goal_scorers_cleaned['log_minute'] = np.log(goal_scorers_cleaned['minute'] + 1)
match_results_cleaned['log_home_score'] = np.log(match_results_cleaned['home_score'] + 1)
match_results_cleaned['log_away_score'] = np.log(match_results_cleaned['away_score'] + 1)
```

```python
# Box plot for log transformed minute in goal_scorers_cleaned
sns.boxplot(x=goal_scorers_cleaned['log_minute'])
plt.title('Box Plot of Log Transformed Goal Scoring Minutes')
plt.xlabel('Log Minute')
plt.show()
```

```python
# Box plot for log transformed home_score in match_results_cleaned
sns.boxplot(x=match_results_cleaned['log_home_score'])
plt.title('Box Plot of Log Transformed Home Scores')
plt.xlabel('Log Home Score')
plt.show()
```

```python
# Box plot for log transformed away_score in match_results_cleaned
sns.boxplot(x=match_results_cleaned['log_away_score'])
plt.title('Box Plot of Log Transformed Away Scores')
plt.xlabel('Log Away Score')
plt.show()
```
